## Absorptance model

Gusarov, A. V., and J-P. Kruth. "Modelling of radiation transfer in metallic powders at laser treatment." International Journal of Heat and Mass Transfer 48.16 (2005): 3423-3434.

```python
from matplotlib import pyplot
%matplotlib inline
import numpy

# Absorptance model - Calculates powder bed effective absorptance based on absorptance of corresponding dense material
def spec_reflection(bulk_refl):
    aspec = numpy.sqrt((1 - bulk_refl))
    Aeffspec = (3*aspec) / (1 + 2*aspec)
    return Aeffspec

def diff_reflection(bulk_refl):
    a = numpy.sqrt(1 - ( (2/3) * bulk_refl ) - ( (1 / 3) * (bulk_refl**2) ) )
    Aeffdiff = (3/4)*((1 - bulk_refl + (3*a))/(1 + 2*a))
    return Aeffdiff

def bulk_reflection(bulk_refl):
    A=1-bulk_refl
    return A

bulk_refl = numpy.linspace(1.0,0.0,200)
dense = bulk_reflection(bulk_refl)

pyplot.figure(figsize=(10,7))
pyplot.rcParams['font.family'] = 'serif'
pyplot.rcParams['font.size'] = 18
pyplot.plot(dense,spec_reflection(bulk_refl),label='Specular')
pyplot.plot(dense,diff_reflection(bulk_refl),'--',label='Diffuse')
pyplot.ylabel('Effective absorptance of powder')
pyplot.xlabel('Absorptance of dense material, 1 - œÅ')
pyplot.xlim(0.0,1.0) ; pyplot.ylim(0.0,1.0)
pyplot.legend(loc='best');
```

## Reflectance at 295K for bulk Bi2Te3 
Reference: Sobolev, V. V., et al. "Reflectivity Spectra of the Rhombohedral Crystals Bi2Te3, Bi2Se3, and Sb2Te3 over the Range from 0.7 to 12.5 eV." physica status solidi (b) 30.1 (1968): 349-355.

```python
Rbulk = 0.5 
Abulk = 1.0 - Rbulk

Aeffop = spec_reflection ( Abulk )
Aeffdiff = diff_reflection ( Rbulk )

print( 'Bi2Te3 powder bed absorptivity - Specular reflection case =', numpy.around(Aeffop,2))
print( 'Bi2Te3 absorptivity powder bed - Diffuse reflection case =', numpy.around(Aeffdiff,2))
```
